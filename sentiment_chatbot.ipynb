```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
# Load sentiment analysis pipeline
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
sentiment_analyzer = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
def analyze_sentiment(user_input):
    # Analyze the sentiment of the user's input
    sentiment = sentiment_analyzer(user_input)[0]
    return sentiment['label'], sentiment['score']  # return label (positive/negative) and confidence score
```

```python
def chatbot_response(user_input):
    sentiment_label, confidence = analyze_sentiment(user_input)

    # Decide the response based on sentiment
    if '5' in sentiment_label:
        return f"Thank you for your positive feedback! I'm glad you're satisfied. (Confidence: {confidence:.2f})"
    elif '1' in sentiment_label or '2' in sentiment_label:
        return f"I'm really sorry to hear you're not happy. How can I assist you further? (Confidence: {confidence:.2f})"
    else:
        return f"I'm here to help. How can I assist you today? (Confidence: {confidence:.2f})"
```

```python
# Simulate user input
user_input = input("User: ")
response = chatbot_response(user_input)
print("Chatbot:", response)
```

```python
def start_chatbot():
    print("Chatbot: Hello! How can I help you today?")

    while True:
        user_input = input("User: ").lower()  # Convert input to lowercase for easier matching

        # Check for exit phrases, including "thank you"
        if user_input in ["exit", "quit", "bye", "thank you", "thanks"]:
            print("Chatbot: You're welcome! Have a great day!")
            break

        response = chatbot_response(user_input)
        print("Chatbot:", response)

# Start the interactive chatbot
start_chatbot()
```
